```julia
using ChargedParticleDynamics
using ChargedParticleDynamics: me, mp, md, mα
```

```julia
import ElectromagneticFields.AxisymmetricTokamakCartesian
AxisymmetricTokamakCartesian.@code()
```

```julia
X₀ = [1.05, 0, 0];
```

```julia
E₀ = 1E6;
```

```julia
α₀ = π/2;
```

```julia
m₀ = md
```

```julia
# Parameters:
#  guiding centre position
#  gyro angle
#  pitch angle
#  energy
#  mass
#  charge number
#  magnetic field unit vectors a(t,x), b(t,x), c(t,x)
#  amplitude of magnetic field B(t,x)

ics = InitialConditions(X₀, 0., α₀, E₀, m₀, 1, a⃗, b⃗, c⃗, B)
```

```julia
charged_particle(ics)
```

```julia
pauli_particle(ics)
```

```julia
guiding_center(ics)
```

```julia
n = 12;
```

```julia
px = zeros(n)
py = zeros(n)
pz = zeros(n)

for i in 1:n
    ics = charged_particle(InitialConditions(X₀, 2π*i/n, α₀, E₀, m₀, 1, a⃗, b⃗, c⃗, B))
    px[i] = ics[1][1]
    py[i] = ics[1][2]
    pz[i] = ics[1][3]
end
```

```julia
scatter([X₀[1]], [X₀[2]], [X₀[3]], xguide="x", yguide="y", zguide="z", aspectratio=1, legend=:none)
scatter!(px, py, pz)
```

```julia
scatter([X₀[1]], [X₀[3]], xguide="x", yguide="z", aspectratio=1, legend=:none)
scatter!(px, pz)
```
